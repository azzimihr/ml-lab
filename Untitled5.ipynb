```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from time import sleep

d = np.array(pd.read_csv('sample_data/mnist_train_small.csv'))
```

```python

m, n = d.shape
np.random.shuffle(d) # shuffle before splitting into dev and training sets

train = d[1000:m].T
Y = train[0]
X = train[1:n]
X = X / 255
_,mt = X.shape

def init():
    global W1, b1, W2, b2
    W1 = np.random.rand(10, 784) - 0.5
    b1 = np.random.rand(10, 1) - 0.5
    W2 = np.random.rand(10, 10) - 0.5
    b2 = np.random.rand(10, 1) - 0.5

def sm(Z):
    return np.exp(Z) / sum(np.exp(Z))

def forward():
    global Z1, A1, Z2, A2
    Z1 = W1.dot(X) + b1
    A1 = np.maximum(Z1, 0)  # Relu
    Z2 = W2.dot(A1) + b2
    A2 = sm(Z2)

def oh():
    ohy = np.zeros((Y.size, Y.max() + 1))
    ohy[np.arange(Y.size), Y] = 1
    return ohy.T

def back():
    global dW1, db1, dW2, db2
    dZ2 = A2 - oh()
    dW2 = dZ2.dot(A1.T) / m
    db2 = np.sum(dZ2) / m
    dZ1 = W2.T.dot(dZ2) * (Z1>0)
    dW1 = dZ1.dot(X.T) / m
    db1 = np.sum(dZ1) / m

def update(lr):
    global W1, b1, W2, b2
    W1 = W1 - lr * dW1
    b1 = b1 - lr * db1
    W2 = W2 - lr * dW2
    b2 = b2 - lr * db2

def accuracy(pred):
    print(pred, Y)
    return np.sum(pred == Y) / Y.size

def gd(lr, iters):
    init()
    for i in range(iters):
        forward()
        back()
        update(lr)
        if i % 25 == 0:
            print(i, accuracy(np.argmax(A2, 0)))
```

```python
gd(0.10, 500)
```
